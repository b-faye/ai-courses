# **1. Overview of Python**

Python is a high-level, **interpreted programming language** known for its simplicity and readability. It was created by Guido van Rossum and first released in 1991. Python supports multiple programming paradigms, including **procedural**, **object-oriented**, and **functional** programming. It has a vast standard library and a vibrant community, making it suitable for various applications, such as **web development**, **data science**, **artificial intelligence**, and **automation**.

image.png

**An interpreted programming language:** is a type of programming language where source code is executed directly by an interpreter, without the need for explicit compilation into machine code beforehand.

**Procedural programming:** is a programming paradigm in which the program's logic is structured around procedures, or routines, consisting of a series of steps executed sequentially to accomplish a task.

**Object-oriented programming:** is a programming paradigm that organizes code around objects, which are instances of classes encapsulating data and behavior, facilitating modular, reusable, and scalable software development.

**Functional programming:** is a programming paradigm focused on composing functions to manipulate data, emphasizing immutability, pure functions, and higher-order functions to facilitate declarative and concise code that avoids side effects.

# **2. Installing Python and setup the environment**

**<h2>Installing Python on Windows:</h2>**

1. **Download Python**:
   - Visit the official Python website: [Python Downloads](https://www.python.org/downloads/).
   - Choose the latest version of Python for Windows.
   - Download the Windows installer (`.exe` file).

2. **Run the Installer**:
   - Once the installer is downloaded, double-click the `.exe` file to run it.
   - Check the box that says "Add Python to PATH" during the installation process. This ensures that Python is added to the system PATH, allowing you to run Python from the command prompt easily.
   - Click "Install Now" and follow the prompts in the installer to complete the installation process.

3. **Verify Installation**:
   - After the installation is complete, open Command Prompt.
   - Type `python --version` and press Enter.
   - You should see the installed Python version displayed.
   - Optionally, you can also type `python` and press Enter to enter the Python interpreter, where you can run Python code interactively.


**<h2>Installing Python on Linux:</h2>**

1. **Open Terminal**:
   - Open your terminal application. You can usually find it in your system's applications menu, or you can press `Ctrl + Alt + T` to open it.

2. **Update Package List**:
   - Run the following command to update the package list:
     ```
     sudo apt update
     ```

3. **Install Python**:
   - Run the following command to install Python:
     ```
     sudo apt install python3
     ```
   - This command will install Python 3, which is the recommended version for most purposes. If you specifically need Python 2 for some reason, you can install it by running `sudo apt install python`.

4. **Verify Installation**:
   - After the installation is complete, you can verify that Python is installed by typing:
     ```
     python3 --version
     ```
   - You should see the installed Python version displayed.



**<h2>Installing Python on macOS:</h2>**

1. **Download Python**:
   - Visit the official Python website: [Python Downloads](https://www.python.org/downloads/).
   - Choose the latest version of Python for macOS.
   - Download the macOS installer (`.pkg` file).

2. **Run the Installer**:
   - Once the installer is downloaded, double-click the `.pkg` file to run it.
   - Follow the prompts in the installer to complete the installation process.

3. **Verify Installation**:
   - After the installation is complete, open Terminal.
   - Type the following command to check the installed Python version:
     ```
     python3 --version
     ```
   - You should see the installed Python version displayed.
