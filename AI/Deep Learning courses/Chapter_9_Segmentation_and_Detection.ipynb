# **Semantic segmentation**

Semantic segmentation is a fundamental task in computer vision that involves partitioning an image into meaningful segments and labeling each segment with a class. Unlike traditional image classification, which assigns a single label to an entire image, semantic segmentation aims to understand the content at a pixel level, providing a more detailed and comprehensive understanding of the visual scene.

image.png

**Key Concepts**

* Pixel-level Classification: Each pixel in an image is classified into a predefined category. For instance, in an image containing a dog, cat, and background, semantic segmentation will label every pixel as either "dog," "cat," or "background."

* Dense Prediction: Unlike tasks such as object detection, which provide bounding boxes around objects, semantic segmentation provides dense predictions, meaning every pixel in the image is assigned a class label.

---

**Applications**

* Autonomous Driving: Identifying and distinguishing between roads, pedestrians, vehicles, and other objects.
* Medical Imaging: Segmenting different anatomical structures or pathological regions in medical scans.
* Aerial Image Analysis: Classifying land cover types, such as vegetation, water bodies, and urban areas.
* Augmented Reality: Enabling real-time interaction with the environment by recognizing and segmenting objects.

---

**Techniques and Models**

* **Fully Convolutional Networks (FCNs):** Introduced by Long et al., FCNs are one of the pioneering architectures for semantic segmentation. They replace the fully connected layers in traditional convolutional neural networks (CNNs) with convolutional layers, enabling the network to make dense predictions.

* **U-Net:** Initially developed for biomedical image segmentation, U-Net has a symmetric encoder-decoder architecture with skip connections that combine high-resolution features from the contracting path with the upsampled output, enhancing the localization capabilities.

* **DeepLab:** The DeepLab series, developed by Google, incorporates atrous (dilated) convolutions and atrous spatial pyramid pooling (ASPP) to capture multi-scale contextual information without losing resolution.

* **Mask R-CNN:** Extending the Faster R-CNN object detection framework, Mask R-CNN adds a branch for predicting segmentation masks for each object instance, combining object detection and segmentation.

* **Transformers:** Recently, transformer-based architectures like Vision Transformers (ViTs) and DETR (Detection Transformer) have shown promising results in semantic segmentation, leveraging self-attention mechanisms to model global relationships in the image.

---

**Challenges**

* **Class Imbalance:** Many segmentation tasks suffer from class imbalance, where some classes (e.g., background) dominate the image, making it difficult for the model to learn minority classes.

* **Computation and Memory:** Semantic segmentation models, especially high-resolution images, require significant computational power and memory.

* **Annotation Effort:** Creating pixel-level annotations is labor-intensive and time-consuming, posing a bottleneck in generating large-scale labeled datasets.

---

**Evaluation Metrics**

**1. Intersection over Union (IoU) / Jaccard Index**

Intersection over Union (IoU) is a measure of the overlap between the predicted segmentation and the ground truth segmentation. It is computed as the ratio of the area of overlap (intersection) between the predicted and ground truth regions to the area of union (sum) of these regions.

Mathematically, IoU is defined as:

$$ IoU = \frac{|A \cap B|}{|A \cup B|} $$

Where:
- $ A $ represents the set of pixels predicted to belong to a certain class.
- $ B $ represents the set of pixels in the ground truth that belong to the same class.
- $ |A \cap B| $ denotes the number of pixels that are correctly classified (true positives).
- $ |A \cup B| $ denotes the total number of pixels that are classified in either $ A $ or $ B $ (true positives + false positives + false negatives).

**2. Pixel Accuracy**

Pixel Accuracy measures the ratio of correctly predicted pixels to the total number of pixels in the image. It gives a straightforward measure of how many pixels are correctly classified across all classes.

Mathematically, Pixel Accuracy (PA) is calculated as:

$$ PA = \frac{\text{Number of correctly classified pixels}}{\text{Total number of pixels}} $$

**3. Mean Intersection over Union (mIoU)**

Mean IoU (mIoU) computes the average IoU across all classes present in the dataset. It provides a balanced evaluation metric, especially useful when dealing with class imbalance, where some classes may be much smaller in area compared to others.

To compute mIoU:
- Calculate the IoU for each class separately.
- Average the IoU values across all classes.

Mathematically, Mean IoU (mIoU) is given by:

$$ mIoU = \frac{1}{N} \sum_{i=1}^{N} IoU_i $$

Where:
- $ N $ is the number of classes.
- $ IoU_i $ is the IoU for class $ i $.

**Summary**

- **IoU**: Measures overlap between predicted and ground truth regions.
- **Pixel Accuracy**: Measures overall accuracy by comparing correctly classified pixels to total pixels.
- **Mean IoU (mIoU)**: Provides an average IoU across all classes, offering a balanced evaluation metric in the presence of class imbalance.

These metrics are crucial in assessing the performance of semantic segmentation models, guiding improvements and comparisons between different approaches.

# **Instance Segmentation**

Instance segmentation is an advanced computer vision task that combines object detection and semantic segmentation. It involves identifying and delineating each distinct object instance within an image, assigning each pixel not only a class label but also a unique identifier corresponding to a specific object instance.

image.png

**Key Concepts**

* **Object-level Segmentation:** Unlike semantic segmentation, which groups pixels into classes (e.g., "person," "car," "tree"), instance segmentation distinguishes between individual object instances of the same class. For example, in an image with multiple dogs, each dog is uniquely identified and segmented.

* **Object Detection + Semantic Segmentation:** Instance segmentation builds upon object detection by adding pixel-level segmentation masks for each detected object instance. This provides a more granular understanding of the spatial extent of each object within the scene.

---

**Applications:**

* **Autonomous Vehicles:** Identifying and tracking multiple pedestrians, vehicles, and other objects simultaneously.
* **Robotics:** Enabling robots to interact with and manipulate objects in complex environments.
* **Retail Analytics:** Counting and tracking individual products on store shelves.
* **Medical Imaging:** Segmenting and analyzing different organs or anomalies within medical scans.
---

**Techniques and Models**

* **Mask R-CNN:** One of the most widely used architectures for instance segmentation, Mask R-CNN extends the Faster R-CNN framework by adding a branch for predicting segmentation masks alongside bounding boxes and class labels. It achieves state-of-the-art performance by integrating object detection and segmentation into a single network.

* **Panoptic Segmentation:** A recent advancement that aims to unify instance segmentation (individual object instances) and semantic segmentation (scene-level classes) into a cohesive framework. Panoptic segmentation assigns a category label and an instance ID to every pixel in an image, combining the strengths of both tasks.

* **Feature Pyramid Networks (FPN):** Often used as the backbone architecture in instance segmentation models, FPNs generate a multi-scale feature pyramid that facilitates accurate localization and segmentation of objects at different scales within the image.

* **PointRend:** An extension to Mask R-CNN that improves the quality of instance segmentation masks by refining predictions at the pixel level using adaptive convolutional operations.

* **Instance Embedding Methods:** Techniques such as MaskTrack R-CNN use instance embeddings to track objects across frames in videos, maintaining consistency in object identities over time.

---

**Challenges**

* **Object Occlusion and Crowding:** Instances of the same class may overlap or occlude each other, making it challenging to accurately delineate boundaries and assign instance IDs.

* **Computational Cost:** Instance segmentation models typically require more computational resources compared to semantic segmentation or object detection due to the additional complexity of predicting segmentation masks for each instance.

* **Dataset Annotation:** Annotating instance segmentation datasets with pixel-level masks for each object instance is labor-intensive and requires meticulous labeling to ensure accurate training.
---

**Evaluation Metrics**

**1. Mean Average Precision (mAP)**

Mean Average Precision (mAP) is a widely used metric in object detection and instance segmentation tasks. It evaluates both the accuracy of object detection (bounding box overlap) and the quality of instance segmentation (mask quality).

**Definition:**

To compute mAP for instance segmentation, we follow these steps:

1. **Precision and Recall Calculation:**
   - **Precision:** The fraction of true positive detections out of all positive detections (both correct and incorrect).
   - **Recall:** The fraction of true positive detections out of all ground truth instances.

2. **Precision-Recall Curve:**
   - For each class, we compute a precision-recall curve by varying the confidence threshold for detection and segmenting objects.
   - Each point on this curve corresponds to a different threshold.

3. **Average Precision (AP):**
   - AP is calculated as the area under the precision-recall curve for each class.
   - AP summarizes the precision-recall curve into a single value, reflecting the overall performance for detecting and segmenting instances of that class.

4. **Mean Average Precision (mAP):**
   - Finally, mAP is the average of AP values across all classes.
   - It provides an aggregate measure of the model's performance across all object classes.

**Mathematical Formulation:**

Let $ \text{TP}(c), \text{FP}(c), \text{FN}(c) $ denote the number of true positives, false positives, and false negatives for class $ c $.

For each class $ c $:

- Precision $ \text{P}(c) = \frac{\text{TP}(c)}{\text{TP}(c) + \text{FP}(c)} $
- Recall $ \text{R}(c) = \frac{\text{TP}(c)}{\text{TP}(c) + \text{FN}(c)} $

Precision-Recall curve is computed by varying the confidence threshold $ \tau $.

Average Precision (AP) for class $ c $ is the area under the precision-recall curve:

$$ \text{AP}(c) = \int_0^1 \text{P}(c)(\text{R}) \, d\text{R} $$

Mean Average Precision (mAP) is the mean of AP values across all classes:

$$ \text{mAP} = \frac{1}{N} \sum_{c=1}^{N} \text{AP}(c) $$

Where \( N \) is the number of classes.

**2. Mask Intersection over Union (Mask IoU)**

Mask IoU measures the spatial overlap between predicted segmentation masks and ground truth masks for individual object instances.

**Definition:**

For each predicted instance $ i $ and corresponding ground truth instance \( j \):

- Let $ M_i $ be the binary mask of predicted instance $ i $.
- Let $ G_j $ be the binary mask of ground truth instance $ j $.

**Intersection over Union (IoU):**
$$ \text{IoU}(M_i, G_j) = \frac{|M_i \cap G_j|}{|M_i \cup G_j|} $$

Where $ |M_i \cap G_j| $ is the number of pixels that are both in $M_i$ and $ G_j $, and $ |M_i \cup G_j| $ is the number of pixels that are in either $ M_i $ or $ G_j $.

**Average Mask IoU:**

To compute the average Mask IoU across all instances, we typically calculate the mean IoU for all predicted instances:

$$ \text{Mean Mask IoU} = \frac{1}{N} \sum_{i=1}^{N} \max_{j} \text{IoU}(M_i, G_j) $

Where $ N $ is the number of predicted instances, and $ \max_{j} \text{IoU}(M_i, G_j) $ denotes the maximum IoU between the predicted mask $ M_i $ and any ground truth mask $ G_j $ of the same class.

**3. Panoptic Quality (PQ)**

Panoptic Quality (PQ) is a metric specifically designed for evaluating panoptic segmentation tasks, which combine both semantic segmentation (assigning class labels to each pixel) and instance segmentation (delineating individual object instances).

**Definition:**

PQ is computed by considering both recognition quality (RQ) and segmentation quality (SQ):

$$ \text{PQ} = \text{PQ}_{\text{RQ}} \times \text{PQ}_{\text{SQ}} $$

Where:

- $ \text{PQ}_{\text{RQ}} $ measures the recognition quality (accuracy of semantic segmentation):
  $$ \text{PQ}_{\text{RQ}} = \frac{\sum_{c} \text{TP}_{\text{seg}}(c)}{\sum_{c} (\text{TP}_{\text{seg}}(c) + \text{FP}_{\text{seg}}(c) + \text{FN}_{\text{seg}}(c))} $$

- $ \text{PQ}_{\text{SQ}} $ measures the segmentation quality (accuracy of instance segmentation):
  $$ \text{PQ}_{\text{SQ}} = \frac{\sum_{i} \text{IoU}(M_i, G_{\pi(i)}) \cdot [|\text{M}_i| \geq \text{V}_{\text{min}}]}{\sum_{i} |\text{M}_i| + \sum_{j} |\text{G}_j|} $$

  Here, $ \text{IoU}(M_i, G_{\pi(i)}) $ is the IoU between the predicted mask $ M_i $ and its matched ground truth mask $ G_{\pi(i)} $.

- $ \text{V}_{\text{min}} $ is a threshold to filter out small instances.

PQ combines both aspects to provide a comprehensive evaluation of the model's performance in panoptic segmentation tasks.


# **Object Detection**

Object detection is a fundamental task in computer vision that involves identifying and localizing multiple objects within an image. Unlike semantic segmentation and instance segmentation, which focus on pixel-level classification and instance-level segmentation respectively, object detection aims to detect and classify objects at a higher level of abstraction, typically using bounding boxes to outline their locations.

image.png

**Key Concepts**

* **Bounding Box Localization:** Object detection algorithms predict bounding boxes that tightly enclose objects of interest within an image. Each bounding box is associated with a class label, indicating the type of object detected (e.g., "person," "car," "dog").

* **Class-specific Detection:** Object detection algorithms are capable of identifying multiple object instances of different classes within a single image, providing a comprehensive inventory of objects present in the scene.

---

**Applications:**

* **Surveillance:** Monitoring crowded environments to detect and track people, vehicles, and suspicious activities.
* **Retail:** Counting and analyzing products on store shelves for inventory management and customer behavior analysis.
* **Natural Disaster Response:** Identifying and locating individuals or vehicles in disaster-stricken areas for rescue operations.
---

**Techniques and Models**

* **Faster R-CNN:** A seminal model in object detection, Faster R-CNN combines region proposal networks (RPN) for generating candidate object regions with a Fast R-CNN backbone for classification and bounding box regression.

* **YOLO (You Only Look Once):** Known for its real-time performance, YOLO approaches object detection as a regression problem to spatially separated bounding boxes and associated class probabilities directly from an input image.

* **Single Shot Multibox Detector (SSD):** SSD predicts object bounding boxes and class probabilities at multiple feature map scales, achieving high detection accuracy and real-time processing speeds.

* **EfficientDet:** Building upon the EfficientNet architecture, EfficientDet scales up object detection models efficiently by optimizing model architecture and training strategies.

* **CenterNet:** CenterNet uses a single-stage detector that directly predicts object centers and their associated bounding boxes and class labels, achieving high accuracy with reduced computational cost.

---

**Comparison with Semantic Segmentation and Instance Segmentation**

**Object Detection vs. Semantic Segmentation:**
* Focus: Object detection identifies and localizes objects at a higher level, providing bounding boxes and class labels without pixel-level segmentation.
* Output: Semantic segmentation labels every pixel in an image with a class label, whereas object detection provides bounding boxes around detected objects.
* Usage: Object detection is suitable for tasks requiring the localization and classification of multiple object instances in a scene, while semantic segmentation is more suited for understanding pixel-wise category information.

**Object Detection vs. Instance Segmentation:**
* Scope: Instance segmentation combines object detection with pixel-level segmentation, providing detailed masks for each instance of detected objects.
* Granularity: Object detection focuses on bounding box localization and classification, while instance segmentation offers finer granularity by segmenting each object instance at the pixel level.
* Applications: Object detection is widely used in scenarios where knowing the presence and approximate location of objects suffices, while instance segmentation is essential when precise delineation of object boundaries is required, such as in medical imaging or robotics.

---

**Challenges**

* **Scale Variation:** Objects can appear at different scales within an image, making it challenging to detect and localize them accurately across different contexts.

* **Occlusion and Clutter:** Overlapping objects or cluttered scenes can obscure parts of objects, affecting the accuracy of detection and localization.

* **Real-time Processing:** Meeting real-time performance requirements while maintaining high detection accuracy remains a significant challenge, especially for applications requiring fast decision-making.
---

**Evaluation Metrics**

* **Intersection over Union (IoU):** Measures the overlap between predicted and ground truth bounding boxes, assessing the spatial accuracy of object localization.

* **Average Precision (AP):** Calculates precision-recall curves to evaluate the detection performance across different confidence thresholds for object detection models.

* **Mean Average Precision (mAP):** The average precision calculated across all object classes, providing a comprehensive evaluation metric for object detection systems.
---

```python
pretrained_model = keras_cv.models.RetinaNet.from_preset(
    "retinanet_resnet50_pascalvoc", bounding_box_format="xywh"
)
```

```python
filepath = tf.keras.utils.get_file(origin="https://www.petinsuranceaustralia.com.au/wp-content/uploads/2014/11/dog-and-cat-1170x777.jpg")
image = keras.utils.load_img(filepath)
image = np.array(image)


visualization.plot_image_gallery(
    np.expand_dims(image, axis=0),
    value_range=(0, 255),
    rows=1,
    cols=1,
    scale=5,
)
```

```python
inference_resizing = keras_cv.layers.Resizing(
    640, 640, pad_to_aspect_ratio=True, bounding_box_format="xywh"
)
image_batch = inference_resizing([image])


class_ids = [
    "Aeroplane",
    "Bicycle",
    "Bird",
    "Boat",
    "Bottle",
    "Bus",
    "Car",
    "Cat",
    "Chair",
    "Cow",
    "Dining Table",
    "Dog",
    "Horse",
    "Motorbike",
    "Person",
    "Potted Plant",
    "Sheep",
    "Sofa",
    "Train",
    "Tvmonitor",
    "Total",
]
class_mapping = dict(zip(range(len(class_ids)), class_ids))


y_pred = pretrained_model.predict(image_batch)
y_pred
```

```python
visualization.plot_bounding_box_gallery(
    image_batch,
    value_range=(0, 255),
    rows=1,
    cols=1,
    y_pred=y_pred,
    scale=5,
    font_scale=0.7,
    bounding_box_format="xywh",
    class_mapping=class_mapping,
)
```
