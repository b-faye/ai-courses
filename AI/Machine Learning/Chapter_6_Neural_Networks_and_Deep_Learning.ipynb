# **Neural Networks**

**<h2>Definition</h2>**

A neural network is a computational model inspired by the way biological neural systems process information. It consists of interconnected layers of nodes (neurons) that simulate the behavior of the human brain. Each neuron receives inputs, processes them using weighted connections and an activation function, and produces an output. Neural networks are capable of learning from data by adjusting the weights through a training process, allowing them to identify patterns and make predictions. They are widely used in various applications, including image and speech recognition, natural language processing, and predictive analytics.

**<h2>Concept of a Neuron</h2>**
- **Biological Inspiration:**
  - Neurons are the basic working units of the brain.
  - They receive inputs, process them, and produce outputs.
- **Mathematical Model:**
  - A simplified model of a neuron can be expressed as:$$y = f\left(\sum_{i=1}^{n} w_i x_i + b\right)$$
  - Where:
    - $ y $ is the output.
    - $ x_i $ are the inputs.
    - $ w_i $ are the weights.
    - $ b $ is the bias.
    - $ f $ is the activation function.

**<h2>Components<h2>**
- **Inputs:**
  - Represent features or data points.
- **Weights:**
  - Parameters that determine the influence of each input.
- **Bias:**
  - An additional parameter that helps the model fit the data better.
- **Activation Function:**
  - Introduces non-linearity into the model.
  - Common functions: Sigmoid, Tanh, ReLU.

**<h2>Network Architectures</h2>**

**1. Single Layer Perceptron (SLP)**
- **Model Representation:**
  - A network with a single layer of output neurons.
  - Output is a weighted sum of inputs passed through an activation function.
- **Mathematical Representation:**
  - For a binary classifier:$$y = f\left(\sum_{i=1}^{n} w_i x_i + b\right)$$
- **Limitations:**
  - Cannot solve non-linearly separable problems (e.g., XOR problem).

image.png

**2. Multilayer Perceptron (MLP)**
- **Structure and Layers:**
  - Consists of an input layer, one or more hidden layers, and an output layer.
  - Each layer is a collection of neurons.
- **Forward Propagation:**
  - Inputs are passed through the network layer by layer.
  - Each layer computes an output using the activation function.
- **Universal Approximation Theorem:**
  - States that a feedforward network with a single hidden layer containing a finite number of neurons can approximate any continuous function.

**<h2>Training Neural Networks</h2>**

**1. Loss Function**
- **Definition and Importance:**
  - Measures the difference between the actual output and the predicted output.
- **Common Loss Functions:**
  - Mean Squared Error (MSE) for regression:$$\text{MSE} = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$$
  - Cross-Entropy for classification:$$\text{Cross-Entropy} = -\frac1N \sum_{i=1}^{N} [y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)]
    $$


**2. Gradient Descent**
- **Concept and Mathematical Formulation:**
  - An optimization algorithm to minimize the loss function.
  - Updates weights iteratively:$$w = w - \eta \frac{\partial L}{\partial w}$$
  - Where $ \eta $ is the learning rate.
- **Variants:**
  - **Batch Gradient Descent:** Uses the entire dataset to compute gradients.
  - **Stochastic Gradient Descent (SGD):** Uses one sample per iteration.
  - **Mini-Batch Gradient Descent:** Uses a small batch of samples per iteration.

**<h2>Mathematical Foundations</h2>**

**1. Linear Algebra**
- **Vectors and Matrices:**
  - Vectors: Ordered set of numbers.
  - Matrices: Rectangular array of numbers.
- **Matrix Multiplication:**
  - Used to compute the output of each layer.
  - If $ A $ is an $ m \times n $ matrix and $ B $ is an $ n \times p $ matrix, the product $ AB $ is an $ m \times p $ matrix.
- **Eigenvalues and Eigenvectors:**
  - Important for understanding properties of matrices used in networks.

**2. Calculus**
- **Differentiation:**
  - Key for understanding how changes in weights affect the loss function.
- **Chain Rule in Backpropagation:**
  - Used to compute gradients efficiently.
  - If $ z = f(y) $ and $ y = g(x) $, then:$$\frac{dz}{dx} = \frac{dz}{dy} \cdot \frac{dy}{dx}$$

# **Introduction to Deeep Learning**

**<h2>Definition and Scope**


- **Deep Learning:**
  - A subset of machine learning where neural networks with many layers (deep networks) are used to model complex patterns in data.
  - Capable of learning hierarchical feature representations from raw data.
- **Difference from Traditional Machine Learning:**
  - Traditional ML often requires manual feature extraction.
  - Deep learning automatically learns features through multiple layers of abstraction.
- **Applications:**
  - Image and speech recognition, natural language processing, autonomous driving, healthcare, and more.


**<h2>Deep Architectures</h2>**

**1. Deep Neural Networks (DNNs)**
- **Layer Depth and Complexity:**
  - DNNs consist of an input layer, multiple hidden layers, and an output layer.
  - Increased depth allows for learning more complex functions.
- **Advantages Over Shallow Networks:**
  - Better at capturing complex patterns.
  - Capable of hierarchical representation learning.



**2. Convolutional Neural Networks (CNNs)**
- **Introduction:**
  - Designed to process data with a grid-like topology, such as images.
  - Utilizes convolutional layers to automatically and adaptively learn spatial hierarchies of features.
- **Mathematical Formulation:**
  - **Convolution Operation:**$$(f * g)(t) = \int_{-\infty}^{\infty} f(\tau) g(t - \tau) \, d\tau$$
  - **Discrete Convolution:**$$(f * g)(i, j) = \sum_m \sum_n f(m, n) g(i - m, j - n)$$
- **Components:**
  - **Convolutional Layers:** Apply convolution operations to the input.
  - **Pooling Layers:** Downsample feature maps to reduce dimensionality and computation.
  - **Fully Connected Layers:** Traditional dense layers that process flattened input from previous layers.
- **Applications:**
  - Image classification, object detection, facial recognition, and more.



image.png

**3. Recurrent Neural Networks (RNNs)**
- **Structure and Working:**
  - Designed for sequential data, where current input depends on previous computations.
  - Maintains a hidden state that captures information from previous time steps.
- **Mathematical Model:**
  - Hidden state update:$$h_t = f(W_h h_{t-1} + W_x x_t + b)$$
  - Output:$$z_t = g(W_y h_t + c)
    $$
- **Challenges:**
  - Difficulty in learning long-term dependencies due to vanishing/exploding gradients.
- **Applications:**
  - Time series prediction, language modeling, speech recognition, and more.



**4. Long Short-Term Memory (LSTM) Networks**
- **Addressing RNN Limitations:**
  - Designed to capture long-term dependencies more effectively.
  - Utilizes gates to control the flow of information.
- **Mathematical Model:**
  - **Forget Gate:**$$f_t = \sigma(W_f \cdot [h_{t-1}, x_t] + b_f)$$
  - **Input Gate:**$$i_t = \sigma(W_i \cdot [h_{t-1}, x_t] + b_i)$$$$\tilde{C}_t = \tanh(W_C \cdot [h_{t-1}, x_t] + b_C)$$
  - **Cell State Update:**$$C_t = f_t \ast C_{t-1} + i_t \ast \tilde{C}_t
$$
  - **Output Gate:**$$o_t = \sigma(W_o \cdot [h_{t-1}, x_t] + b_o)$$$$h_t = o_t \ast \tanh(C_t)$$
- **Applications:**
  - Similar to RNNs but more effective for long-term dependencies.

**5.Transformers**
- **Introduction:**
  - Designed for handling sequential data without recurrence.
  - Uses self-attention mechanisms to capture dependencies between distant tokens.
- **Mathematical Formulation:**
  - **Self-Attention:**$$\text{Attention}(Q, K, V) = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V$$
  - **Multi-Head Attention:**$$\text{MultiHead}(Q, K, V) = \text{Concat}(\text{head}_1, \text{head}_2, \ldots, \text{head}_h)W^O$$
- **Components:**
  - **Encoder-Decoder Structure:**
    - Encoder processes the input sequence.
    - Decoder generates the output sequence.
  - **Positional Encoding:** Adds positional information to input tokens.
- **Applications:**
  - Natural language processing tasks like translation, summarization, and language modeling.

**<h2>Libraries</h2>**

**1. TensorFlow**
- **Introduction:**
  - Developed by Google.
  - Provides a comprehensive, flexible ecosystem for building deep learning models.
- **Features:**
  - Keras API for high-level model building.
  - TensorFlow Extended (TFX) for production ML pipelines.


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist

# Load and preprocess the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 784).astype('float32') / 255
x_test = x_test.reshape(-1, 784).astype('float32') / 255

# Define a simple MLP
model = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test accuracy: {accuracy}')

# Predict on new data
predictions = model.predict(x_test)

# Example: Print the prediction for the first test sample
print(f'Predicted label for the first test sample: {tf.argmax(predictions[0]).numpy()}')
print(f'True label for the first test sample: {y_test[0]}')
```

**2. PyTorch**
- **Introduction:**
  - Developed by Facebook's AI Research lab.
  - Known for its dynamic computation graph and ease of use.
- **Features:**
  - Strong support for GPU acceleration.
  - Integration with Python for deep integration with the Python ecosystem.


```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Define the SimpleMLP model
class SimpleMLP(nn.Module):
    def __init__(self):
        super(SimpleMLP, self).__init__()
        self.fc1 = nn.Linear(784, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 10)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.softmax(self.fc3(x), dim=1)
        return x

# Load and preprocess the MNIST dataset
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])

train_dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
test_dataset = datasets.MNIST(root='./data', train=False, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Instantiate the model, loss function, and optimizer
model = SimpleMLP()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training the model
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        data = data.view(data.size(0), -1)  # Flatten the input data
        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()

    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {loss.item()}')

# Evaluating the model
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for data, target in test_loader:
        data = data.view(data.size(0), -1)
        outputs = model(data)
        _, predicted = torch.max(outputs.data, 1)
        total += target.size(0)
        correct += (predicted == target).sum().item()

print(f'Test Accuracy: {100 * correct / total:.2f}%')

# Making predictions
model.eval()
with torch.no_grad():
    for data, target in test_loader:
        data = data.view(data.size(0), -1)
        outputs = model(data)
        _, predicted = torch.max(outputs.data, 1)
        print(f'Predicted: {predicted[0].item()}, True: {target[0].item()}')
        break  # Print prediction for the first sample and exit the loop
```
