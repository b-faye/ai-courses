**keywords:**

Sure, here are the definitions for each relevant keyword in the context of deep learning:

1. **Deep Supervision Branch:**
   - In deep learning, a deep supervision branch refers to an additional set of layers or pathways in a neural network that facilitates the training process by providing intermediate supervision signals. These signals help guide the learning process in deeper layers of the network, potentially improving convergence and preventing vanishing gradients (ex. residual connection).

2. **Feature Pyramid:**
   - A feature pyramid is a hierarchical representation of features extracted from an input image at multiple scales. In deep learning, feature pyramids are commonly used in tasks such as object detection and semantic segmentation to detect objects or features of different sizes across various spatial resolutions.

3. **Auxiliary Supervision:**
   - Auxiliary supervision, also known as auxiliary loss or auxiliary task, is a technique used in deep learning where additional supervised signals are incorporated into a neural network to guide the training process. These auxiliary signals are typically derived from intermediate layers of the network and are used to supervise the learning of features at those layers.

4. **Reversible Architecture:**
   - In deep learning, a reversible architecture refers to a network design where the operations performed during forward propagation can be easily reversed during backward propagation. This allows for efficient memory usage and enables the network to reconstruct input data from intermediate representations, which can be beneficial for tasks such as gradient computation and parameter updates.

5. **Integration Network:**
   - An integration network is a component within a neural network architecture that combines or integrates information from multiple sources. In the context of the provided text, the integration network is inserted between the layers of the feature pyramid hierarchy and the main branch of the network to combine gradients from different prediction heads.

6. **Gradient:**
   - In deep learning, gradients refer to the partial derivatives of the loss function with respect to the model parameters. Gradients indicate the direction and magnitude of parameter updates required to minimize the loss function during training through techniques such as gradient descent.

7. **Parameter Updates:**
   - Parameter updates refer to the process of adjusting the weights and biases of a neural network based on computed gradients. During training, the gradients are used to update the parameters of the network in the direction that minimizes the loss function, gradually improving the network's performance.

8. **Inference:**
   - Inference in deep learning refers to the process of using a trained model to make predictions or decisions on new, unseen data. During inference, input data is passed through the trained model, and the model's output is computed without any further parameter updates or training iterations.

**Abstract**

This paper addresses the importance of designing appropriate objective functions and architectures in deep learning to ensure accurate prediction results. It highlights the issue of information loss during data transmission through deep networks due to layer-by-layer feature extraction and spatial transformation. The concept of programmable gradient information (PGI) is introduced to address this challenge, enabling complete input information for calculating objective functions and obtaining reliable gradient information to update network weights. Additionally, a new lightweight network architecture, Generalized Efficient Layer Aggregation Network (GELAN), is proposed based on gradient path planning. GELAN, utilizing PGI, demonstrates superior results on lightweight models compared to existing methods. Experimental validation on the MS COCO dataset for object detection confirms that GELAN, using conventional convolution operators, achieves better parameter utilization than state-of-the-art methods based on depth-wise convolution. PGI is versatile and applicable to various model sizes, enabling better performance for train-from-scratch models compared to pre-trained models on large datasets.

# **wav2vec 2.0: A Framework for Self-Supervised Learning of Speech Representations [Link](https://arxiv.org/abs/2006.11477)**

* Methods like CLIP learn powerful image representations from text.

**Abstract**

* Learning powerful representations from speech audio alone followed by fine-tuning on transcribed speech can outperform the best semi-supervised methods.

* wav2vec 2.0 masks the speech input in the latent space and solves a contrastive task defined over a quantization of the latent representations which are jointly learned.

---------------------------------------------------------------------------

**Introduction**

* **Self-supervised learning** has emerged as a paradigm to learn general data representations from unlabeled examples and to fine-tune the model on labeled data.

* **wav2vec 2.0** is a self-supervised learning of representations from raw audio data.

* wav2vec 2.0 jointly learning **discrete speech units** with **contextualized representations** achieves substantially better results than fixed units learned in a prior step.

* **keywords:**<br>
  * CTC loss : Connectionist Temporal Classification
  * WER : Word Error Rate

---------------------------------------------------------------------------

**Model**

Model is composed of:
  * **Feature Encoder: Temporal Convolution Network (TCN)**
  * **Transformer: Contextualized Representations**
  * **Quantization Module**

**1. Feature Encoder**<br>

$$f : X \rightarrow Z$$

where $X$ is input raw audio and $Z = z_1, ..., z_T$ output latent speech representations for $T$ time-steps.

The encoder consists of several blocks containing a temporal convolution followed by layer normalization and a GELU activation function. The total stride of the encoder determines the number of time-steps $T$ which are input to the Transformer.


**2. Transformer**

$$f : Z \rightarrow C$$
where $C = c_1, ..., c_T$, allows to capture information from the entire sequence.

* Relative positional embeddings (with a convolutional layer) is used instead of fixed positional embeddings.

* The output of the convolution layer followed by GELU is added to inputs, and then apply Layer Normalization.

**3. Quantization Module**

For self-supervised training, the output of the **feature encoder** is discretized to a finite set of speech representations via **product quantization** using **Gumbel Softmax**.

Given $G$ codebooks (sequences) with $V$ entries ($V$ sequence) $e \in \mathbf{R}^{V \times G/d}$, all codebooks. We choose one entry (vector) from each codebook(squence) and concatenate resulting vectors $e_1$, $e_2$, ..., $e_G$ (each $e_i$ is vector containing $G/d$ elements), and apply linear transformation to obtain $q$.

The Gumbel softmax enables choosing discrete codebook entries in a fully differentiable way. The feature encoder output $z$ is mapped to $l \in \mathbf{R}^{G\times V}$ logits and the probabilities for choosing the
$v$-th codebook entry for group $g$ are
$$p_{g, v} = \frac{e^{(l_{g,v} + n_v)}/\tau}{\sum_{k=1}^{V} e^{(l_{g,k} + n_k)}/\tau}
$$

where $\tau$ is a non-negative temperature, $n = - \log(- \log(u))$ and $u$ are uniform samples from $\mathcal{U}(0, 1)$. $v \in \{1 ... V\}$.

During the forward pass, codeword $i$ is chosen by $i = argmax_j p_{g,j}$ ($j \in \{1, ..., V\}$) and in the backward pass, the true gradient of the Gumbel softmax outputs is used.

---------------------------------------------------------------------------

**Training**

* **Masking**<br>
  We mask a proportion of the feature encoder outputs, or time steps before feeding them to the context network and replace them with a trained feature vector shared between all masked time steps; we do not mask inputs to the quantization module.

* **Objective**<br>
  * **Contrastive loss** $\mathcal{L}_m$: Learn representations of speech audio between masked representations and quantized latent speech representation.
  * **Code diversity loss**$\mathcal{L}_d$: to encourage the model to use the codebook entries equally often.
  $$\mathcal{L}_m + \alpha\mathcal{L}_d$$
  where $\alpha$ is a tuned hyperparameter.

* **Contrastive loss**<br> Given context output (Transformer) $c_t$, the model needs to identify the true quantized latent speech representation $q_t$ in a set of $K+1$ candidate representations $\tilde{q} \in Q_t$ which includes $q_t$ and $K$ distractors. Distractors are uniformly sampled from other masked time steps of the same utterance.
$$\mathcal{L}_m = - \log \frac{\exp(\text{sim}(c_t, q_t)/\kappa)}{\sum_{\tilde{q} \sim Q_t} \exp(\text{sim}(c_t, \tilde{q})/\kappa)}
$$

* **Diversity Loss**<br>
The contrastive task depends on the codebook to represent both positive and
negative examples and the diversity loss $\mathcal{L}_d$ is designed to increase the use of the quantized codebook representations.

We encourage the equal use of the $V$ entries in each of the $G$ codebooks by maximizing the entropy of the averaged softmax distribution $l$ over the codebook entries for each codebook $\bar{p}_g$ across a batch of utterances; the softmax disribution does not contain the gumbel noise
nor a temperature.
$$\mathcal{L}_d = \frac{1}{G} \sum_{g=1}^{G} -H(\bar{p}_g)  = \frac{1}{G} \sum_{g=1}^{G} \sum_{v=1}^{V} \bar{p}_{g,v} \log \bar{p}_{g,v}
$$

---------------------------------------------------------------------------

**Fine-tuning**

Pre-trained models are fine-tuned for speech recognition by adding a randomly initialized linear projection on top of the **context network** into **C** classes representing the vocabulary of the task.
