There are four fundamental types of fusion [Vision based body gesture meta features for Affective Computing](https://arxiv.org/abs/2003.00809):

**1. Early fusion or feature level fusion:** Early fusion is a technique in data analysis where features from different modalities are combined at the input level before processing or analysis.

**2. Intermediate fusion or intermediate level fusion:** Intermediate fusion refers to a method in data analysis where features from multiple modalities are combined at an intermediate stage of processing, typically after initial feature extraction but before final decision-making or analysis.

**3. Late fusion or decision level fusion:** Late fusion is a method in data analysis where features from different modalities are processed separately and then combined at a later stage, often during decision-making or analysis, to integrate information from diverse sources.

**4. Hybrid fusion or hybrid level fusion:** uses the combination of feature level fusion and decision level fusion, in consideration of advantages of both these fusion strategies.  In  [Amer, et al. (2018)](https://scholar.google.com/scholar_lookup?title=Deep%20multimodal%20fusion%3A%20A%20hybrid%20approach&publication_year=2018&author=M.R.%20Amer&author=T.%20Shields&author=B.%20Siddiquie&author=A.%20Tamrakar&author=A.%20Divakaran&author=S.%20Chai), the authors proposed a hybrid approach as deep multimodal fusion to classify sequential data from multiple modalities.

# **Alignment**

Alignment, in the context of using different modalities, refers to the process of bringing different sources of information into a coherent and synchronized representation. When working with multiple modalities, such as text, images, audio, or video, alignment involves ensuring that the information from each modality is correctly matched and synchronized to capture the underlying relationships or semantics across modalities.

In the context of Vision-Language Models (VLMs), there several methods to build it:

* **Contrastive learning.**
* **Masked language-image modeling.**
* **Encoder-decoder modules with transformers and more.**

These architectures can learn complex relations between the various modalities and provide state-of-the-art results.

## **Contrastive Learning**

Contrastive learning is a self-supervised learning approach that aims to train a model by maximizing the similarity between representations of positive pairs (e.g., similar images) while minimizing the similarity between representations of negative pairs (e.g., dissimilar images).

##**Encoder-decoder modules with transformers and more**

**<h2>1. PrefixLM</h2>**

PrefixLM is an NLP learning technique mostly used for model pre-training. It inputs part of the text (a prefix) and learns to predict the next word in the sequence. In Visual Language Models, PrefixLM enables the model to predict the next sequence of words based on an image and its respective prefix text. It leverages a Vision Transformer (ViT) that divides an image into a one-dimensional sequence of patches, where each patch represents a local image region.

Then, the model applies convolution or linear projection over the processed patches to generate contextualized visual embeddings. For text modality, the model converts the text prefix relative to the patch into a token embedding. The encoder-decoder blocks of the transformer receive both visual embedding and token embedding. It is there that the model learns the relationships between the embeddings.

**SimVLM** is a popular architecture utilizing the PrefixLM learning methodology. It has a simpler Transformer architecture than its predecessors, surpassing their results in various benchmarks. It uses a transformer encoder to learn image-prefix pairs and a transformer decoder to generate an output sequence. The model also demonstrates good generalization and zero-shot learning capabilities.

**<h2>2. VirTex</h2>**
VirTex uses a convolutional neural network to extract image features and a textual head with transformers to manage text prefixes. You can train the model end-to-end to predict the correct image captions by feeding image-text pairs to the textual head.

While PrefixLM techniques require you to train visual and textual encoders from scratch, Frozen PrefixLM allows you to use pre-trained networks and only update the parameters of the image encoders.

## **6. MDETR**

**MDETR is DETR where encoder take concatenation of image and text sequence (instead of image only), and add two loss functions**

**Abstract**

* **Limit of existing methods** :
    * Multi-modal reasoning systems rely on a pretrained object detector to extract regions of interest from the image.

    * These object detector are trained independently of the downstream task and on a fixed vocabulary of objects and attributes.

    * This makes it challenging for such systems to capture the long tail of visual concepts expressed in free form text.


* **MDETR approach** : Use transformer-based architecture to reason jointly over text and image by fusing (concatenation) the two modalities at an early stage of the model.

---------------------------------------------------------------------------

**Introduction**

* Existing multi-modal object detections use **frozen detector** that are pretrained in a fixed vocabulary.

* MDETR is a end-to-end multi-modal reasoning based on DETR.

---------------------------------------------------------------------------

**Method**

* **DETR component**

  * a backbone (CNN)
  * transformer encoder-decoder
  * the output of the backbone is used as input of the encoder after flattening to get sequence representation (H $\times$ W, C).
  * The decoder takes as input a set of $N$ **learned embeddings called object queries**. For cross-attention, key and value are encoder embeddings. The number of queries acts as the number of objects that the model can detect simultanously (no object $\emptyset$ is used).
  * The FFN of the detector gives **box coordinates and class labels**.
  * DETR is trained **Hungarian matching loss**
  * **bipartite matching** is computed between the $N$ proposed objects and the ground-truth objects.
  * Each matched object is supervised using the corresponding target as ground-truth, while the un-matched objects are supervised to pre-
  dict the "no object" label $\emptyset$.
  * The classification head is supervised using standard **cross-entropy**.
  * The bounding box head is supervised using a combination of **absolute error (L1 loss)** and **Generalized IoU**.

* **MDETR component**
  * Image is encoded by a CNN backbone then flattened. To conserve the spatial information, a 2-D positional embeddings are added to this flattened vector.
  * Text is encoded by a pretrained transformer language model.
  * Modality dependent linear projection to both the image and text features to project them into a shared embedding space.
  * These feature vectors are then concatenated on the sequence dimension to yield a single sequence of image and text features.
  * A transformer encoder called **cross encoder** uses the concatenated sequence to better representation.
  * DETR (only the decoder) is used for **predicting the actual boxes**.


* **Training**

  MDETR add two loss functions to encourage alignment between the image and the text.

  * **Soft token prediction** : is based on token spans after tokenization using a BPE scheme. This loss allows to extends detection in several tokens not only one token. Any query that is not matched to a target is trained to predict the “no object” label ∅. Note that several words in the text could correspond to the same object in the image, and conversely several objects could correspond to the same text. For example, “a couple” referred to by two boxes in the image, could further be referred to individually in the same caption.

  * **Contrastive alignment** : While the soft token prediction
  uses positional information to align the objects to text, the
  contrastive alignment loss enforces alignment between the
  embedded representations of the object at the output of the
  decoder, and the text representation at the output of the cross
  encoder.

  * After matching, the total loss consists of the box
  prediction losses (L1 & GIoU), soft-token prediction loss (soft cross entropy, use set of tokens not one token),
  and the contrastive alignment loss.

## **7. Grounded Language-Image Pre-training (GLIP)**

**Abstract**
* There are two types of models:
    * detection model: uses fixed set of classes
    * grounded model: uses detection and grounded data

* GLIP allows model for learning:
    * object-level
    * language aware
    * semantic-rich visual representations

* GLIP learns from both **detection** and **Grounded data** (association data).
* GLIP generates **grounded boxes** in self-training fashion.
* GLIP is pre-trained on 27M grounded data:
    * 3M human-annotated
    * 24M web-crawled **image-text pairs**
* GLIP after pre-training can be used on:
    * zero-shot object-level recognition tasks.
    * few-shot object-level recognition tasks.

---------------------------------------------------------------------------

**Introduction**

* **Phrase grounding**: task for identifying fine-grained correspodance between **phrases in sentence** and **objects (regions)** in an image.

* GLIP unifies **phrase grounding** and **object detection**

* GLIP takes as input **image** and **text prompt** that describes all categories in image.

* In COCO, a text prompt contains **80 phrases** (categories or class), i.e., the 80 COCO object class names, joined by ". ": Person. Bicycle... Hairdryer.

* **Any detection model** can be converted to a **grounded model** by replacing the **object classification logits** in its box classifier with the **word alignment scores**, i.e., dot product of the region (or box) visual features and the token (or phrase) language features.

* **Technical detail**: CLIP fuses vision and language only at the **last dot product layer**, GLIP use CLIP approach but add **deep cross-modality fusion** in middle (early fusion), that allow to learn **high quality language-aware visual representations** and to **achieve superior transfer learning performance**.

image.png

**Related work**

* **Prase grounding** = **word-to-region matching**

* GLIP uses **Dynamic Head: Unifying Object Detection Heads with Attentions** for region detection.

* **ViLD** is proposed for **zero-shot** object detection, by distilling the knowledge from CLIP/ALIGN.

* **MDETR** trains and end-to-end model on existing **multi-modal datasets explicit alignment between phrases in text and objects in image.**

**Grounded Language Image Pre-training**

* **Object detection**<br>
model feeds an **input image** into visual encoder $Enc_I$ with CNN or Transformer as backbone and extracts region/box features $O$.<br>
Each region/box features is fed into **two prediction heads**: **box classifier $C$** and **box regressor $R$**, which are trained with classification loss $\mathcal{L}_{cls}$ and the localisation loss $\mathcal{L}_{loc}$, respectively:$$\mathcal{L} = \mathcal{L}_{cls} + \mathcal{L}_{loc}$$ <br>

**In two-stage detectors**, a separate region proposal network (RPN) with RNP loss $\mathcal{L}_{RPN}$ is used to distinguishforeground and background. It is merged into  $\mathcal{L}_{loc}$. In **one-stage detectors**, localization loss $\mathcal{L}_{loc}$ may also contain the **centerness loss**.

The **box classifier $C$** is a **simple linear layer**, and the classification loss $\mathcal{L}_{cls}$ can be written:
$$
O = Enc_I(Img)
$$
$$
S_{cls} = OW^T
$$
$$
\mathcal{L}_{cls} = loss(S_{cls};T)
$$
where $O \in \mathbf{R}^{N\times d}$ are the object/region/box feature of the input image. $W \in \mathbf{R}^{c\times d}$ is the weight matrix of the box classifier $C$, $S_{cls} \in \mathbf{R}^{N\times c}$ are the output classification logits and $T \in \{0, 1\}^{N\times c}$ is the target matching between regions and classes computed from the classical **many-to-1 matching** or **bipartie Hungarian match**.<br>
$loss(S;T)$ is **cross-entropy for two-stage detectors** and **focal loss for one-stage detectors**.
<br>
<br>

* **Object detection as phrase grounding**<br>
In grounding model, the alignment scores $S_{ground}$ between **image regions** and **words** in the prompt is computed:
$$
O = Enc_I(Img)
$$
$$
P = Enc_L(prompt)
$$
$$
S_{ground} = OP^T
$$
where $P \in \mathbf{R}^{M\times d}$ is contextual word/token features from the language encoder and plays a similar role to the weight matrix $W$. The number of (sub)-word tokens $M$ is always larger than the number of phrases $c$ in the text prompt due to four reasons:
    * some phrases contains multiple words, e.g., "traffic light".
    * some single-word phrases are splitted into multiple (sub)-word tokens, e.g., “toothbrush” to “tooth#” and “#brush”.
    * some are the added tokens, such as “Detect:”, “,”, special tokens in language models.
    * a [NoObj] token is added at the end of the tokenized sequence.

When the $loss$ is a (focal) **binary sigmoid** loss, orginal matrix $T \in \{0, 1\}^{N\times c}$ is expanded by $T' \in \{0, 1\}^{N\times M}$ by making all sub-words positive match if phrase is a positive match and all added tokens negative match to all image features. The loss remains the same $loss(S_{ground}; T')$. **During inference, we
average token probabilities as the phrase probability.**

---------------------------------------------------------------------------

**Language-aware Deep Fusion**

* GLIP uses two fusion mechanism:
    * early fusion: cross-attention fusion
    * late fusion: alignment

* **DyHead **: Image encoder
* **BERT**: Text encoder
* The deep-fused encoder is:
$$
O_{t2i}^i, P_{i2t}^i = \text{X-MHA}(O^i, P^i), i \in \{0, 1, .., L - 1\}
$$
$$
O^{i+1} = \text{DyHeadModule}(O^i + O^{t2i}), O = O^{L}
$$
$$
P^{i+1} = \text{BERTLayer}(P^i + P_{i2t}^i), P = P^{L}
$$

where $L$ is the number of DyHeadModules in DyHead, BERTLayer is newly-added BERT Layers on top of the pre-trained BERT, $O^0$ denote the visual features from the vision backbone, and $P^0$ denote the token features from the language backbone (BERT). The cross-modality communication is achieved by the cross-modality multi-head attention
module (X-MHA), followed by the single modality fusion and updated in. Without added context vectors ($O_{t2i}^i$
for vision modality and $P_{i2t}^i$ for language modality), the model is reduced to a late-fusion model.

* In the cross-modality multi-head attention module (X-MHA), each head computes the context vectors of one modality by attending to the other modality:
$$
O^{(q)} = OW^{(q,I)}, \quad P^{(q)} = PW^{(q,L)}, \quad \text{Attn} = \frac{O^{(q)}(P^{(q)})^T}{\sqrt{d}}\\
P^{(v)} = PW^{(v, L)}, O_{t2i} = SoftMax(Attn)P^{(v)}W^{(out, I)}\\
O^{(v)} = OW^{(v, I)}, P_{t2i} = SoftMax(Attn^T)O^{(v)}W^{(out, L)}
$$

The deep-fused encoder brings two benefits. 1) It improves the phrase grounding performance. 2) It makes the learned visual features language-aware, and thus the model's prediction is conditioned on the text prompt.

## **8. Contextual Object Detection**

**abstract**

* LLM can be used in vision:
    * Image Captioning (IC)
    * Visual Question Answering (VQA)
* In those tasks, models **lack the essential perception ability**, i.e., **object detection**: **Location**, **Identification** and **association** visual objects with language inputs for **human-AI interaction**.

* ContextDet proposes **03 scenarios**:
    * Language Cloze Test: Complete text prompt using image (mask some words)
    * Image captioning
    * Visual question answering

* ContextDet gives **2 outputs**:
    * **Text**
    * **image with bouding boxes**

* ContextDet use **03 submodels**:
    * **Visual Encoder**: extracts visual representations
    * **Pre-trained LLM**: for multimodal contextual decoding
    * **Visual Decoder**: predicts bouding boxes given contextual object words

* ContextDet uses the paradigm: **generate-then-detect**

image.png

**Introduction**

* There are two detection categories:
    * **Clozed-vocabulary detection**: identifying and localizing a predefined set of object classes (i.e., YOLO, RCNN, DETR, etc) use paradigme **detect-then-classify**.
    * **Open-vocabulary detection**: ability of a system to identify and classify objects, even those not explicitly defined in a fixed set (GLIP, MDETR, etc).

* **Open-vocabulary object detection** can improve the zero-shot transfer ability for novel classes, they are constrained by the scale of pre-defined
novel categories, making them incapable of detecting objects for a human language vocabulary.

* To address, ContextDet propose a novel framework: **generate-then-detect**.

* Submodels in ContextDet:
    * **Visual Encoder**: produces **local** and full visual tokens
    * **Pre-trained LLM**: perfoms text generation conditionned by **local visual tokens** and **task-related language tokens**.
    * **Visual decoder**: consists of **multiple cross-attention layers**, taking LLM tokens as **queries** and full visual tokens as **keys** and **values**.

**Related work**

* **Visual Encoder**<br>
For image input $x \in \mathbf{R}^{3\times H\times W}$ a frozen vision backbone $ f_\phi$ is used to extract **image-level spatial features** : $v = f_\phi(x) \in \mathbf{R}^{d\times h\times w}$. $v$ is then transformed with two distinct representations:<br>
    * **Local Visual tokens**: adaptive average pooling with $p$ bins is applied on $v$, followed by a linear projection then flattened to 1D: $z = \text{AvgPool(Linear($v$))} \in \mathbf{R}^{d_1\times p}$, where $d1$ is input dimension of the LLM.

    * **Full Visual tokens**: The 2D spatial feautres of $v$ if flattened in 1D: $m = h \times w$ tokens. The output is used as input by 6 Transformer layers $f\psi$ to encode full visual tokens: $c = f\psi(v) \in \mathbf{R}^{d_2\times m}$

* **LLM** <br>
    * **Multimodal tokens**: The inputs of the LLM consist of 1) the local visual tokens **$z  \in \mathbf{R}^{d_1\times p}$** and 2) a serie of language tokens $t_{1:l} = \{t_1, ..., t_l\} \in \mathbf{R}^{d_1\times l}$.
    
    * **Multimodal Prefixed LLM Decoding**: a pre-trained LLM $\theta$ can be conditionned on a prefix $w_{1:n}$ that contains multimodal tokens to generate text in autoregressive way:
    $$
    p(w_{n+1:L} | w_{1:n}) = \prod_{i=n+1}^{L} p_{\theta}(w_{i+1} | w_{1:i}).
    $$
    
    Here, the prefix $w_{1:n} = [z, t_{1:l}] \in \mathbf{R}^{d_1\times (p+l)}$ is obtained via concatenating the **local visual tokens** and **language tokens**.<br>
    LLM consists of multiple Transformer layers (TransLayer) and final Feed Forward Network (FFN). To generate new token: $$e_{n+1} = TransLayer(w_{1:n}),$$ then, the FFN computes the probability distribution: $$p(w_{n+1}) = softmax(FFN(e_{n+1})).$$

* **Visual Decoder** ContextDet unlike common **detect-then-classify**, use **generate then detect** paradigm.<br>

    * **Contextual LLM Tokens as Conditional Object Queries**: prefix $t_{1:l}$ and generated tokens $w_{n+1:L}$ are used to predict a **binary classification probability of noun object words**. Then automatically select those language tokens related to object words (e.g., "bridge", "groom", "dog") as contextual object tokens and take their latent embeddings as conditional inputs for the visual decoder.<br>
    $N$ (the number of fixed boxes) learnable object queries $q$ are used. For contextual token, e.g., "bridge", the conditionnal object queries linked to "bridge", is obtained by incorporating the corresponding latent embedding $e$, from the LLM with the object queries:
    $$ \bar{q} = q + Linear(Repeat(e)).$$
    Here the latent embedding $e$ is repeated $N$ times so as to align with the number of object queries

    * **Conditional Multimodal Context Decoding**: To model cross-modal contextual relationships, **6 Transformers cross-attention layers** is used: $$\hat{q} = CrossAttentionLayers(c, \bar{q})$$ By doing so, the visual decoder learns to focus on specific areas of the visual context that are relevant to the conditional query for "bride".

    * **Box and Matching Predictions for Contextual Words.**: The binary matching and box prediction are computed from $\hat{q}$: $$p = FFN_{cls}(\hat{q}) \in \mathbf{R}^{N\times 2}$$ $$b = FFN_{box}(\hat{q}) \in \mathbf{R}^{N\times 4}$$ Where $p$ is to probability of being matched or not matched given the conditional object word, and $b$ indicates the predicted box cordinates.

    * **Conditional Matching for Label Assignment** a conditionnal modification is done in the **optimal bipartie matching in DETR** that finds the best match between the set of $N$ predictions and the set of ground truth objects. In ContextDET, only the ground-truth bounding box that match the conditional object words are involved in loss computation. This conditional matching ensures that the model focuses solely on the objects described by the language queries.

**Training Details**

**Multi-scale deformable attention** and **IoU-based** label assignment are used to accelerate the convergence speed.

- ContextDet use the conditional matching to derive the label assignments, which include ground-truth matching labels $\hat{p}$ and the associated box cordinates $\hat{b}$.

- For language predicted token $w$, ContextDet gets the ground truth token $\hat{w}$ by using method like cloze test.

- $\bar{w}$ indicates token belongs on an object word or not.

$$
\mathcal{L} = \lambda_{\text{cls}} \mathcal{L}_{\text{cls}}(p, \hat{p}) + \lambda_{\text{box}} \mathcal{L}_{\text{box}}(b, \hat{b}) + \lambda_{\text{lm}} \mathcal{L}_{\text{lm}}(w, \hat{w}) + \lambda_{\text{noun}} \mathcal{L}_{\text{noun}}(w, \bar{w})
$$

1. $\mathcal{L}_{\text{cls}}$ is a binary softmax classification loss of two classes: matched vs. not matched.

2. $\mathcal{L}_{\text{box}}$ is the box-related loss is either L1 loss or GIoU loss.

3. $\mathcal{L}_{\text{lm}}$ is softmax classification loss over the vocabulary size $\mathcal{W}$ of the LLM Tokenizer.

4. $\mathcal{L}_{\text{noun}}$ is a binary classification loss that determines whether a token is an object word or not.
